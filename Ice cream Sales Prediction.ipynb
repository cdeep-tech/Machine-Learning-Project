```python
# Get the data

import pandas as pd
import numpy as np

df = pd.read_csv("SalesData.csv")
df
```

```python
""" Data preproceeing Steps
 1. Data Cleaning
 2. Split the data into train & test"""
```

```python
# Data CLeaning

df.isnull().sum()
```

```python
# Seperate Features & Label

X=df[["Temperature"]]    # Features
Y=df["Revenue"]          # Label
```

```python
X
```

```python
Y
```

```python
# Train & test Data splitting
from sklearn.model_selection import train_test_split

x1,x2,y1,y2 = train_test_split(X,Y,
                 test_size=0.2,
                 random_state=2)
```

```python
# Train the model

from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x1,y1)
```

```python
# Test the Model

yp = model.predict(x2)
yp
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,10))
plt.scatter(x2, yp)
plt.scatter(x2,y2)
```

```python
model.score(x2,y2) * 100
```
